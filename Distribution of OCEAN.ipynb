```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
psycho_data = pd.read_csv("D:/new/psychometric_info.csv")
bins=max(psycho_data["O"])-min(psycho_data["O"])+1
plt.hist(psycho_data["O"],bins)
plt.xlabel('O value')
plt.ylabel('number of people')
plt.show()
```
